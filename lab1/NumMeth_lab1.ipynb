# 1. Линейная интерполяция  
*Для линейной интерполяции между двумя скалярными величинами a и b используют одну из двух следующих функций.  
Используя формулы Голдберга, докажите, что*
- *первая функция дает точный результат при a==b и может дать неточный результат при t==1.f;*
- *вторая функция дает точный результат при t==1.f и может дать неточный результат при a==b.*  
##########################################################    
### a) lerp_1
#### a1.
$r = a + t \times (b - a)$  
$b \ominus a = (1 + \delta_1)(b - a)$  
$t \otimes (b \ominus a) = (1 + \delta_2)(1 + \delta_1)t(b - a) = 0$ при $a == b$  
$a \oplus t \otimes (b \ominus a) = a \oplus 0 = a$ при $a == b$  
#### a2.
Если $a \neq b$, то погрешность будет ненулевая, даже если $t == 1.f$  
Умножение на $1.f$ вроде происходит без погрешности, но это не спасает:  
$a \oplus (b \ominus a) = a \oplus (1 + \delta_1)(b - a) = (1 + \delta_2)(a + (1 + \delta_1)(b - a)) \neq a + 1.f \times (b - a)$ при $a \neq b$
